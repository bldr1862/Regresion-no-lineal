```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Data import

```python
df = pd.read_csv('datosT2.txt', sep = ' ', header=None)
df.columns = ['month', 'value']
fig, ax = plt.subplots()
ax.plot(df.month,df.value)
ax.set_title('Cantidad de pasajeros vs Meses')
ax.set_xlabel('Meses')
ax.set_ylabel('Cantidad de pasajeros')
plt.show()
```

### Augmented dataset

```python
df_augmented = pd.DataFrame()
df_augmented['ones'] = np.ones(df.shape[0])
df_augmented['first_order'] = df.month.apply(lambda x: x**1)
df_augmented['second_order'] = df.month.apply(lambda x: x**2)
df_augmented['third_order'] = df.month.apply(lambda x: x**3)
df_augmented['forth_order'] = df.month.apply(lambda x: x**4)
df_augmented['value'] = df.value
```

### Train test split

```python
y = df_augmented.value
X = df_augmented.drop(labels = 'value', axis = 1)

df_train = df_augmented.iloc[0:109,:]
df_test = df_augmented.iloc[109:,:]

X_train_1_order = np.matrix(df_train.iloc[:,0:2])
X_train_2_order = np.matrix(df_train.iloc[:,0:3])
X_train_3_order = np.matrix(df_train.iloc[:,0:4])
X_train_4_order = np.matrix(df_train.iloc[:,0:5])
y_train = np.transpose(np.matrix(df_train.value))

train_set = [X_train_1_order,X_train_2_order,X_train_3_order,X_train_4_order]
```

```python
params = []
ro = 5
for t_set in train_set:
    p = np.zeros(5)
    den = np.dot(np.transpose(t_set), t_set) + ro*np.identity(t_set.shape[1]) 
    num = np.dot(np.transpose(t_set), y_train)
    p_cal = np.dot(np.linalg.inv(den),num)
    for i,v in enumerate(p_cal):
        p[i] = v
    params.append(np.array(p))
```

```python
columns = ['first_o','second_o','third_o','forth_o']

for p,col in zip(params,columns):
    df_test[col] = df_test.iloc[:,0:5].apply(lambda x: np.sum(np.multiply(x,p)) , axis = 1)
    df_train[col] = df_train.iloc[:,0:5].apply(lambda x: np.sum(np.multiply(x,p)) , axis = 1)
```

```python
fig, ax = plt.subplots()
color = ['b','r','g','y','c']
labels = ['real values','first order','second order', 'third order', 'forth order']
for col, c, label in zip(['value']+columns, color, labels):
    ax.plot(df_test.first_order, df_test[col], color = c, label = label)
ax.set_title('Resultados de regresiones')
ax.set_xlabel('Meses')
ax.set_ylabel('Pasajeros')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
for i,col in enumerate(columns):
    t1 = np.sum(np.square(df_train['value']-df_train[col]))
    t2 = ro*np.linalg.norm(params[i])
    print("Polinomio Order {}".format(i+1), "Valor de la funcion de costo {}".format(t1+t2))
    # Polinomio de segundo orden es el minimo de los minimos
```

### Polinomio cuadratico + seno modulado por exponencial

```python
df_train = df.iloc[0:109,:]
df_test = df.iloc[109:,:]
params_pol = params[1][0:3] # parametros modelo segundo order
```

```python
from scipy import optimize
def modelo1(x, teta1,teta2,teta3,teta4):
    pol_term =params_pol[0] + params_pol[1]*x + params_pol[2]*x**2
    sin_term = teta1*np.sin(teta2*x+teta3)*np.exp(teta4*x)
    return pol_term+sin_term

def objective(teta):
    return np.sum(np.square(df_train.value - modelo1(df_train.month, *teta)))


teta_start = (0.1, 0.5, 0.5, 0.01)
sol = optimize.minimize(objective, teta_start, method = 'BFGS')
params_sin_1 = sol.x # agregar a la lista los nuevos parametros
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(df_train.month, df_train.value, 'r', lw=2)
ax.plot(df_train.month, modelo1(df_train.month, *sol.x), 'b', lw=2)
plt.axvline(x=109)
ax.plot(df_test.month, df_test.value, 'r', lw=2)
ax.plot(df_test.month, modelo1(df_test.month, *params_sin_1), 'b', lw=2)
ax.set_title('Estimacion Maxima Verosimilitud')
plt.show()
```

### Segunda sinusoide

```python
def modelo2(x, teta1,teta2,teta3,teta4):
    pol_term =params_pol[0] + params_pol[1]*x + params_pol[2]*x**2 
    sin_term_1 = params_sin_1[0]*np.sin(params_sin_1[1]*x+params_sin_1[2])*np.exp(params_sin_1[3]*x)
    sin_term_2 = teta1*np.sin(teta2*x+teta3)*np.exp(teta4*x)
    return pol_term+sin_term_1+sin_term_2

def objective(teta):
    return np.sum(np.square(df_train.value - modelo2(df_train.month, *teta)))

teta_start = (1.05, 1.05, 1.05, 0.01)
sol = optimize.minimize(objective, teta_start, method = 'BFGS')
params_sin_2 = sol.x # agregar a la lista los nuevos parametros
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(df_train.month, df_train.value, 'r', lw=2)
ax.plot(df_train.month, modelo2(df_train.month, *sol.x), 'b', lw=2)
plt.axvline(x=109)
ax.plot(df_test.month, df_test.value, 'r', lw=2)
ax.plot(df_test.month, modelo2(df_test.month, *params_sin_2), 'b', lw=2)
ax.set_title('Estimacion Maxima Verosimilitud')
plt.show()
```

### Error de prediccion

```python
def error_prediccion(y_true,y_aprox):
    error = np.sum(np.square(y_true-y_aprox))
    return error
```

```python
error_modelo_1 = error_prediccion(df_test.value, modelo1(df_test.month, *params_sin_1))
error_modelo_2 = error_prediccion(df_test.value, modelo2(df_test.month, *params_sin_2))
print("Error modelo 1: {}".format(error_modelo_1))
print("Error modelo 2: {}".format(error_modelo_2))
```
